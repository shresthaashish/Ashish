```python
def cleaner(Band,upper,lower):
    m=len(Band)
    for index in range(0, m):
        if  Band[index]>lower:
            if  Band[index]>upper: 
                Band[index]=0
        else:
            Band[index]=0
    return Band 
```

```python
def read_pl_rsr(rsr_file):

    with open(rsr_file, 'rb') as csvfile:
        data = list(csv.DictReader(csvfile, delimiter='\t'))
    # wavelength units: um
    pl_wavelength = [float(row['wavelength (um)']) for row in data]

    def t_str_to_t(t_str):
        # transmission is recorded as a percent string (e.g. '50.0%')
        return (float(t_str[:-1]) / 100)

    def normalize(t):
        return (np.array(t).astype(np.float)) / max(t)

    pl_rsr = [normalize([t_str_to_t(row['R']) for row in data]),
              normalize([t_str_to_t(row['G']) for row in data]),
              normalize([t_str_to_t(row['B']) for row in data])]

    assert len(pl_wavelength) == \
        len(pl_rsr[0]) == \
        len(pl_rsr[1]) == \
        len(pl_rsr[2])

    return (pl_wavelength, pl_rsr)
```
